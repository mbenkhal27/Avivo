```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd
Location_one=r'/Users/mbenkhal/Documents/Avivo/MergedData.xlsx'
Data = pd.read_excel(Location_one, 'Sheet1')

Data['Status'] = "false"
Data['Expected Hours'] = ''
for index, row in Data.iterrows():
    if(row[10] == "Completed program" or row[10] == "Transferred to other program"):
        
        Data['Status'][index] = "true"
    
        
Data.head(5)
```

```python

for index , row in Data.iterrows():
    if(row[1] == "Treatment - Relapse Men"):
        Data['Expected Hours'][index] = '150'
    elif(row[1] == "Treatment - Family - Outpatient"):
        Data['Expected Hours'][index] = '250'
    elif(row[1] == "Treatment - IDD Women"):
        Data['Expected Hours'][index] = '200'
    elif(row[1] == "Treatment - IDD Men"):
        Data['Expected Hours'][index] = '200'
    else:
        Data['Expected Hours'][index] = '150'
```

```python
sum_hours_program = Data.groupby(['Program','Status']).sum()

sum_hours_program['number_hours'].head(5)
```

```python
count = Data.groupby(['Program','Status']).count()
count['Client_number'].head(5)
```

```python
df = pd.DataFrame(columns= ["Program","Status","Avg_hours","Expected_Hours"])
df['Avg_hours']= sum_hours_program['number_hours']/count['Client_number']

names = list(df.index)
names
```

```python
for i in range(0, df.shape[0]):
    temp = ' '.join(names[i])
    if("Treatment - Relapse Men" in temp):
        df['Expected_Hours'][i] = '150'
        df['Program'][i] = "Treatment - Relapse Men"
        if("true" in temp):
            print(temp)
            df['Status'][i] = 'completed'
        elif("false" in temp):
            print("not"+ temp)
            df['Status'][i] = 'not completed'
            
    elif("Treatment - Family - Outpatient" in temp):
        df['Expected_Hours'][i] = '250'
        df['Program'][i] = "Treatment - Family - Outpatient"
        if("true" in temp):
            df['Status'][i] = 'completed'
        elif("false" in temp):
            df['Status'][i] = ' not completed' 
    elif("Treatment - IDD Women" in temp):
        df['Expected_Hours'][i] = '200'
        df['Program'][i] = "Treatment - IDD Women"
        if("true" in temp):
            df['Status'][i] = 'completed'
        elif("false" in temp):
            df['Status'][i] = ' not completed' 
    elif("Treatment - IDD Men" in temp):
        df['Expected_Hours'][i] = '200'
        df['Program'][i] = "Treatment - IDD Men"
        if("true" in temp):
            df['Status'][i] = 'completed'
        elif("false" in temp):
            df['Status'][i] = ' not completed' 
    else:
        df['Expected_Hours'][i] = '150'
        df['Program'][i] = "Treatment - Relapse Women"
        if("true" in temp):
            df['Status'][i] = 'completed'
        elif("false" in temp):
            df['Status'][i] = 'not completed' 
df
```
